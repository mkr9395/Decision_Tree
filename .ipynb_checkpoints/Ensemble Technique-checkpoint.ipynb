### Types of Ensemble Learning:

1. **Voting Ensemble** - Base Models have different algorithm getting same data. All models have same weightage.


2. **Bagging** - Bootstrap Aggregation. All models have same algorithm but are given different data.
    
    - Random Forest - Base models are Decision tree, given different set of data.


3. **Boosting** - Mistakes are passed on to the next model.

    - Adaboosting
    
    - Gradient Boosting
    
    - XgBoost


4. **Stacking** - Similar as Voting Ensemble but more wight is given to the model making most correct predictions.

#### Benefits of Ensemble Technique

1. Improvement in Performance


2. Reduces errors i.e reduces bias and variance. 

    - Better models should have low bias, low variance.
    

3. Robustness

## Voting Ensemble

In Voting Ensemble for Regression: We pass the data through all the models and then take mean of all the outputs.

In Voting Ensemble for Classification: We take the most occuring output through all the models as final output.

#### Why voting works:

1. Models should be dissimilar and independent. Similar model altogether won't give good results.

2. Accuracy of each model should be minimum 50%.

Voting Classifier : https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.VotingClassifier.html
        
Voting Regressor:  https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.VotingRegressor.html

### Voting Types:


1. Hard



2. Soft

### For Voting Classification 

```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import sklearn
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
df = pd.DataFrame(iris.data)
```

```python
df.head()
```

```python
df.columns = iris.feature_names
```

```python
df.head()
```

```python
df['Species'] = iris.target
```

```python
df.head()
```

```python
import seaborn as sns
sns.pairplot(df,hue='Species')
```

```python
# to make classification task difficult we will take only 2 columns i.e Sepal_length and Sepal_width

new_df = df[df['Species']!=0][['sepal length (cm)','sepal width (cm)','Species']]
new_df.head()
```

```python
new_df.shape
```

```python
X = new_df.iloc[:,0:2]
X.head()
```

```python
y = new_df.iloc[:,-1] 
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
```

```python
clf1 = LogisticRegression()
clf2 = RandomForestClassifier()
clf3 = KNeighborsClassifier()
```

```python
# creating estimator objects which are list of tuples('string', estimator)

estimators = [('lr',clf1),('rf',clf2),('knn',clf3)]
```

```python
# cross validation on each estimator
for estimator in estimators:
    
    x = cross_val_score(estimator[1],X,y,cv=10,scoring='accuracy')
    print(estimator[0],np.round(np.mean(x),2))

# getting accuracy of each model
```

```python
from sklearn.ensemble import VotingClassifier
```

### Hard Voting 

```python
vc = VotingClassifier(estimators = estimators,voting='hard')
x = cross_val_score(vc,X,y,cv=10,scoring='accuracy')
print(np.round(np.mean(x),2))

#getting accuracy of hard voting
```

### Soft Voting

```python
vc1 = VotingClassifier(estimators = estimators, voting ='soft')
x = cross_val_score(vc1,X,y,cv=10,scoring='accuracy')
print(np.round(np.mean(x),2))

# getting accuracy of Soft voting
```

### Weighted Voting

giving different weightage to different models

```python
for i in range(1,4):
    for j in range(1,4):
        for k in range(1,4):
            
            vc = VotingClassifier(estimators=estimators, voting = 'soft',weights = [i,j,k])
            x = cross_val_score(vc,X,y,cv=10,scoring='accuracy')
            print("for i={}, j={}, k={}".format(i,j,k),np.round(np.mean(x),2))
```

best accuracy "72%" is when i=logistic which has 75% accuracy is given more weightage and j= random_forest and k = knn are given equal weights

### Classifier for Same Algorithm

```python
from sklearn.svm import SVC
```

```python
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples=1000, n_features=15, n_redundant=5, random_state=2)
```

```python
svm1 = SVC(probability=True, kernel='poly', degree=1)
svm2 = SVC(probability=True, kernel='poly', degree=2)
svm3 = SVC(probability=True, kernel='poly', degree=3)
svm4 = SVC(probability=True, kernel='poly', degree=4)
svm5 = SVC(probability=True, kernel='poly', degree=5)


estimators = [('svm1',svm1),('svm2',svm2),('svm3',svm3),('svm4',svm4),('svm5',svm5)]

for estimator in estimators:
    
    x = cross_val_score(estimator[1],X,y, cv=10, scoring = 'accuracy')
    print(estimator[0],np.round(np.mean(x),2))
    
    
```

```python
# putting all 5 in voting classifier for soft

vc1 = VotingClassifier(estimators=estimators,voting='soft')

x = cross_val_score(vc1,X,y, cv=10, scoring ='accuracy')
print(np.round(np.mean(x),2))
```

```python
# putting all 5 in voting classifier for hard

vc1 = VotingClassifier(estimators=estimators,voting='hard')

x = cross_val_score(vc1,X,y, cv=10, scoring ='accuracy')
print(np.round(np.mean(x),2))
```

### for Voting Regressor

```python
from sklearn.datasets import load_boston
```

```python
X,y = load_boston(return_X_y=True)
```

```python
X.shape
```

```python
y.shape
```

```python
X
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.model_selection import cross_val_score
```

```python
lr = LinearRegression()
dt = DecisionTreeRegressor()
svr = SVR()
```

```python
estimators = [('lr',lr),('dt',dt),('svr',svr)]
```

```python
for estimator in estimators:
    
    scores = cross_val_score(estimator[1],X,y, scoring='r2',cv=10)
    print(estimator[0],np.round(np.mean(scores),2))
```

```python
from sklearn.ensemble import VotingRegressor
```

```python
vr = VotingRegressor(estimators)

scores = cross_val_score(vr,X,y,scoring='r2',cv=10)
print("Voting Regressor",np.round(np.mean(scores),2))
```

##### experimenting with weights

```python
for i in range(1,4):
    for j in range(1,4):
        for k in range(1,4):
            
            vr = VotingRegressor(estimators,weights=[i,j,k])
            
            scores = cross_val_score(vr,X,y,scoring='r2',cv=10)
            
            print("For i={}, j={}, k={}".format(i,j,k),np.round(np.mean(scores),2))
    
```

best accuracy is for i=2, j=1, k=1  = 46%

### Regressor for Same Algorithm

```python
dt1 = DecisionTreeRegressor(max_depth=1)
dt2 = DecisionTreeRegressor(max_depth=2)
dt3 = DecisionTreeRegressor(max_depth=3)
dt4 = DecisionTreeRegressor(max_depth=4)
dt5 = DecisionTreeRegressor(max_depth=5)
```

```python
estimators = [('dt1',dt1),('dt2',dt2),('dt3',dt3),('dt4',dt4),('dt5',dt5)]
```

```python
for estimator in estimators:
    
    scores = cross_val_score(estimator[1],X,y,scoring='r2',cv=10)
    
    print(estimator[0],np.round(np.mean(scores),2))
```

```python
vr = VotingRegressor(estimators)
scores = cross_val_score(vr,X,y,scoring='r2',cv=10)
print("Voting regressor",np.round(np.mean(scores),3))
```
