## Classification

```python
clas = tree.DecisionTreeClassifier(max_depth=5)
iris = load_iris()
```

```python
X_train = iris.data
y_train = iris.target
```

```python
X_train
```

```python
clas.fit(X_train,y_train)
```

```python
clas.get_params()
```

```python
from sklearn.tree import plot_tree
```

```python
plot_tree(clas)
```

### now using dtreeviz

```python
viz = dtreeviz(clas,
               X_train,
               y_train,
               feature_names = iris.feature_names, # input column
               class_names = ['setosa','versicolor','virginica'], #output column
               scale = 2)
viz
```

## Regression

```python
regr = tree.DecisionTreeRegressor(max_depth=2)

boston = load_boston()
```

```python
X_train = boston.data
y_train = boston.target
```

```python
regr.fit(X_train,y_train)
```

```python
viz = dtreeviz(regr,
              X_train,
              y_train,
              target_name='price',
              feature_names=boston.feature_names,
               scale =1.5
              )
```

```python
viz
```

## Horizontal Decision Tree

```python
clas = tree.DecisionTreeClassifier()  
iris = load_iris()

X_train = iris.data
y_train = iris.target
clas.fit(X_train, y_train)


X = iris.data[np.random.randint(0, len(iris.data)),:]


viz = dtreeviz(clas, 
               X_train,
               y_train,
               target_name='price',
               feature_names=iris.feature_names, 
               class_names=["setosa", "versicolor", "virginica"], 
               scale = 1.5,
               orientation='LR')
viz
```

### 4. Show prediction path

```python
clas = tree.DecisionTreeClassifier()  
iris = load_iris()

X_train = iris.data
y_train = iris.target
clas.fit(X_train, y_train)

X = iris.data[np.random.randint(0, len(iris.data)),:]

viz = dtreeviz(clas, 
               X_train,
               y_train,
               feature_names=iris.feature_names, 
               class_names=["setosa", "versicolor", "virginica"],
               scale=1.5,
               X=X)
viz
```
