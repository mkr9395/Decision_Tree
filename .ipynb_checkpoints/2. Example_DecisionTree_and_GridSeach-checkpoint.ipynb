### Post Pruning  -  Removing some branches via ccp_alphas

Seeing the results and making changes

```python
path = dtc.cost_complexity_pruning_path(X1,y1) #known as ccp
```

```python
path

# ccp_alpha= 0 means a full grown tree
```

```python
ccp_alphas = path.ccp_alphas # passing key as alphas is dictionary type
```

```python
ccp_alphas
```

```python
# training and testing model on basis of all the ccp value
dtc_model2 = []

for ccp in ccp_alphas:
    
    dtc_m = DecisionTreeClassifier(ccp_alpha=ccp)
    dtc_m.fit(X1_train,y1_train)
    dtc_model2.append(dtc_m)
```

```python
dtc_model2 

# from same datset we are getting 8 models
```

##### need to select the ccp_alpha value where accuracy of train and test is same which will prevent overfitting

```python
# going through all models

train_score = [i.score(X1_train,y1_train) for i in dtc_model2]
train_score

# getting training score for all 8 models.
```

```python
# going through all models

test_score = [i.score(X1_test,y1_test) for i in dtc_model2]
test_score

# getting training score for all 8 models.
```

```python
fig, ax = plt.subplots()
ax.set_xlabel("ccp_alphas")
ax.set_ylabel("accuracy")
ax.set_title("Accuracy vs alpha for new training and new testing sets")
ax.plot(ccp_alphas, train_score, marker='o', label = 'train_new', drawstyle = 'steps-post')
ax.plot(ccp_alphas, test_score, marker='o', label = 'test_new', drawstyle = 'steps-post')
ax.legend()
plt.show()
```

### GridSearchCV

Decison Tree Classifier Scikit-learn: https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html

1. **Max_Depth** : no of levels of division. Zyada hone pe Overftting ho skta hai. Kam hone pe underfitting.


2. **Min Sample Split :** we can control ki split tabhi hoga jab minimum itne number of rows bache hain uss node k upar.
    
    - jitns value increse hoga chances of underfitting and jitna ghatega utna overfitting


3. **Max Features :** Next node ko maximum kitne columns milenge to decide further splittining.


4. **Max Leaf Nodes:** Total itne hi Leaf nodes banenge. (Leaf node is point where no further splitting can be done)

    - value badha toh overfitting, ghata toh under
